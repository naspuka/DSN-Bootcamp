# Data Preparation

```python
from data_utils.utils import get_data

data = get_data('data',GRAY=True)
```

```python
random.shuffle(data)
X = []
y = []
Y = []
for features, label in data:
    X.append(features)
    Y.append(label)
cloth_mapping = {'not_yoruba':0, 'yoruba':1}

for i in Y:
    y.append(cloth_mapping.get(i))
X = np.array(X)
X = utils.np_utils.normalize(X, axis=1)
y = np.array(y)
```

```python
clothDistribution = Counter(list(y))
print("The data contains {} cloth images of  non yoruba tribe and {} cloth images of yoruba tribe".format(clothDistribution[0],clothDistribution[1]))
```

```python
print('The shape of the features', X.shape)
print('The shape of the lable', y.shape)
```

```python
num_training = int(0.8 * X.shape[0])
num_test = X.shape[0] - num_training

mask = range(num_training, num_training + num_test)
X_test = X[mask]
y_test = y[mask]

mask = range(num_training)
X = X[mask]
y = y[mask]

print('Train data shape: ', X.shape)
print('Train labels shape: ', y.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
```

```python
encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)

X = X.reshape(X.shape[0], 64, 64, 1)

y = utils.np_utils.to_categorical(encoded_Y)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape =(64,64,1) ))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dropout(0.2, noise_shape=None, seed=None))

model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
filepath = "weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=100)
# Fit the model
model.fit(X, y, validation_split=0.30, epochs=500, batch_size=10, callbacks=[checkpoint,es], verbose=0)
```

```python
X_test = X_test.reshape(X_test.shape[0], 64, 64, 1)
predict = model.predict([X_test])
y_test_pred = np.argmax(predict,axis=1)
```

```python
names =list(cloth_mapping)
```

```python
def plot_confusion_matrix(cm, names, title='Confusion Matrix', cmap=plt.cm.Blues):
    from matplotlib.pylab import rcParams
    rcParams['figure.figsize'] = 5,5
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(names))
    plt.xticks(tick_marks, names, rotation=45)
    plt.yticks(tick_marks, names)
    fmt =  'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),horizontalalignment="center",color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

```python
cm = confusion_matrix(y_test, y_test_pred)
plot_confusion_matrix(cm, names)
```

```python
accuracy = accuracy_score(y_test, y_test_pred)
print('Accuracy on test data',accuracy )
```
